```python
!pip install beautifulsoup4
from bs4 import BeautifulSoup
import requests

URL = "https://www.cricbuzz.com/cricket-match/live-scores"
soup = BeautifulSoup(requests.get(URL).content,"html.parser")
mtch_list_from_page = soup.find_all(class_="cb-mtch-lst")
```

```python
mtch_dict = []
for mtch in mtch_list_from_page:
    match={}
    data = mtch.find_all(class_="cb-lv-scr-mtch-hdr")
    match["name"]=data[0].find("a").text  
    match["id"]=data[0].find("a")["href"].split("/")[2]
    mtch_dict.append(match)
```
